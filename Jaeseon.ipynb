```python
import tensorflow as tf
from tensorflow.keras.applications.vgg16 import VGG16
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras import layers
from tensorflow.keras.layers import Dense, Dropout, Input
from tensorflow.keras.models import Model
```

```python
# 모델 불러오기
vgg_model = VGG16(input_shape=(224,224,3), weights='imagenet', include_top=False)
```

```python
vgg_model.summary()
```

```python
vgg_model.trainable=True
global_average_layer = tf.keras.layers.GlobalAveragePooling2D()

model = tf.keras.Sequential([
    vgg_model,
    global_average_layer,
    layers.Dense(19, activation='softmax'),

])
```

```python
model.summary()
```

```python
model.compile(loss='categorical_crossentropy',
                     optimizer='adam',
                     metrics=['accuracy'])
```

```python
train_dir ='/content/drive/MyDrive/crime_project/images/train'

image_generator = ImageDataGenerator( rescale=1./255)

# 데이터 구조 생성
train_data_gen = image_generator.flow_from_directory(batch_size=16,
                                                           directory=train_dir,
                                                           subset='training',
                                                           shuffle=True,
                                                           target_size=(224, 224),
                                                           class_mode='categorical')

# valid_data_gen = image_generator.flow_from_directory(batch_size=16,
#                                                          directory=train_dir,
#                                                          subset='validation',
#                                                          target_size=(224, 224),
#                                                          class_mode='categorical')
```

```python
valid_data_gen = image_generator.flow_from_directory(batch_size=16,
                                                         directory=train_dir,
                                                         subset='validation',
                                                         target_size=(224, 224),
                                                         class_mode='categorical')
```

```python
# 모델 학습
history = model.fit(train_data_gen, epochs=1, validation_data=valid_data_gen)
```
